### Loading Data


```python
df1=pd.read_csv(r"C:\Users\Nagaraj Nagappan\Desktop\MLFlow2\Inventory.csv")
df1 = df1.rename(columns={'Inventory Date Time': 'Date'})
```

```python
# Convert the 'datetime_column' to datetime format
df1['datetime_column'] = pd.to_datetime(df1['Date'])

# Extract only the date and create a new column 'date_column'
df1['New_Date'] = df1['datetime_column'].dt.date

df1.head()
```

```python
df1 = df1.drop(columns=['Date'])
```

```python
df2=pd.read_csv(r"C:\Users\Nagaraj Nagappan\Desktop\MLFlow2\Sale.csv")
df2=df2.rename(columns={'Sales  Date Time': 'Date'})

```

```python
df2['datetime_column'] = pd.to_datetime(df2['Date'])

df2['New_Date'] = df2['datetime_column'].dt.date

df2.head()
```

```python
df2 = df2.drop(columns=['Date'])
```

```python
df3=pd.read_csv(r"C:\Users\Nagaraj Nagappan\Desktop\MLFlow2\Store Product Delivery.csv")
df3 = df3.rename(columns={'Delivered Date Time': 'Date'})
```

```python
df3['datetime_column'] = pd.to_datetime(df3['Date'])

df3['New_Date'] = df3['datetime_column'].dt.date

df3.head()
```

```python
df3 = df3.drop(columns=['Date'])
```

```python
concatenated_df = pd.concat([df1, df2, df3], ignore_index=True, sort=False)
concatenated_df
```

```python
concatenated_df.isnull().sum()
```

```python
numeric_columns = ['Gross Volume', 'Net Volume', 'Temperature', 'Gallons', 'OrderItemId']
concatenated_df[numeric_columns] = concatenated_df[numeric_columns].apply(lambda x: x.fillna(x.mean()))  # You can use median instead of mean if you prefer

# Impute categorical columns
categorical_columns = ['Store', 'Product']
concatenated_df[categorical_columns] = concatenated_df[categorical_columns].apply(lambda x: x.fillna(x.mode()[0]))

# Verify that there are no more missing values
print(concatenated_df.isnull().sum())
```

```python
concatenated_df=concatenated_df.dropna()
```

```python
print(concatenated_df.isnull().sum())
```

```python
concatenated_df.to_csv("Joined_Data.csv", index=False)
```
